```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import random

# Persiapan Data
num_candidates = 4
candidates = ['Candidate {}'.format(i+1) for i in range(num_candidates)]
votes = [0] * num_candidates

# Inisialisasi Grafik
fig, ax = plt.subplots()
bars = ax.bar(candidates, votes)
ax.set_ylim(0, 100)  # Sesuaikan rentang sumbu y

# Fungsi untuk menginisialisasi animasi
def init():
    for bar in bars:
        bar.set_height(0)
    return bars

# Fungsi untuk melakukan pembaruan animasi
def update(frame):
    # Simulasi pemungutan suara acak
    for i, bar in enumerate(bars):
        votes[i] = random.randint(0, 100)
        bar.set_height(votes[i])
    return bars

# Membuat animasi
ani = animation.FuncAnimation(fig, update, frames=range(100),
                              init_func=init, blit=True, interval=100)

# Menampilkan animasi sebagai video HTML
HTML(ani.to_jshtml())
```
